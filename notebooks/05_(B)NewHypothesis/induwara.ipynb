## 1: Reliability Check for Survey 
verify the reliability of the survey questions for Perceived Ease of Use (PEOU) (Section 3, Q13-22) and Intention to Purchase Online (Section 8, Q56-59). Reliability ensures that the questions consistently measure the intended constructs.
### 1.1 Cronbach’s Alpha

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('../../data/cleaned/cleaned_survey_data.csv')  # Replace with your CSV file path

# Define question columns
peou_questions = ['peou_navigation_1', 'peou_navigation_2', 'peou_learning_1', 'peou_learning_2', 'peou_instructions_1', 'peou_instructions_2', 'peou_error_1', 'peou_error_2', 'peou_response_1', 'peou_response_2']  # PEOU
intention_questions = ['opi_satisfaction', 'opi_behavior_change', 'opi_convenience', 'opi_value']  # Intention to Purchase Online

# Function to calculate Cronbach's Alpha
def cronbach_alpha(items):
    n = items.shape[1]  # Number of items
    item_vars = items.var(axis=0, ddof=1).sum()  # Sum of item variances
    total_var = items.sum(axis=1).var(ddof=1)  # Variance of total scores
    return (n / (n - 1)) * (1 - item_vars / total_var)

# Calculate Cronbach's Alpha for PEOU
peou_alpha = cronbach_alpha(df[peou_questions])
print(f"Cronbach's Alpha for PEOU: {peou_alpha:.3f}")

# Calculate Cronbach's Alpha for Intention
intention_alpha = cronbach_alpha(df[intention_questions])
print(f"Cronbach's Alpha for Intention: {intention_alpha:.3f}")
```
